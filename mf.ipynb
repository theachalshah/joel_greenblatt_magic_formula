# Joel Greenblatt's Magic Formula

```python
import pandas as pd
```

Import scrapped data from any free source

```python
df = pd.read_excel('mf.xlsx')
```

```python
df
```

```python
df = df.dropna(how='all')
```

```python
df
```

Rank them by rank of ROIC + rank of EARNINGS_YIELD (Cheap+Good)

```python
df['ey_rank'] = df['EARNINGS_YIELD'].rank(ascending=True)
df['roc_rank'] = df['ROIC'].rank(ascending=True)
df['magic_rank'] = (df['ey_rank'] + df['roc_rank']).rank()
```

```python
df
```

Sort and Invest in top 30 companies for a year 

```python
mf = df.sort_values(by = 'magic_rank', ascending=False)
```

```python
mf.to_csv('mf_july.csv', index=False)
```
